```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
```

```python
uni_data = pd.read_excel("data_insight_file.xlsx", skiprows=24, header=0)
```

```python
uni_data.head(30)
```

```python
birm_df = uni_data[(uni_data.INSTITUTION_CODE=="B32")]
```

```python
birm_df = birm_df.drop(columns=["inst_all", "INSTITUTION_CODE"])
birm_df.head()
```

```python
total_df = birm_df[(birm_df.statistic=="All placed applicants") & (birm_df.equality_dimension=="Total")]
```

```python
total_df
```

```python
eighteens = total_df[(total_df.agegroup=="18 year olds")]['value']
```

```python
all_ages = total_df[(total_df.agegroup=="All ages")]['value']
mature = []
chart_size = 8000
```

```python
for i in range(0, eighteens.size):
    mature.append((-eighteens[eighteens.keys()[i]]+all_ages[all_ages.keys()[i]]))
```

```python
mature
```

```python

#present data as line graph
x = total_df[(total_df.agegroup=="18 year olds")]['Cycle']
y1 = total_df[(total_df.agegroup=="18 year olds")]['value']
y2 = total_df[(total_df.agegroup=="All ages")]['value']


y3 = mature

plt.figure(figsize=(13,8))
plt.plot(x, y2, color="green", label="All ages")
plt.plot(x, y1, color="orange", label="18 year olds")
plt.plot(x,y3, color="grey", label="Other students")




# Create a legend for our chart



plt.grid()
plt.xlabel("Year")
plt.ylabel("Student numbers")
plt.title(f"change in total student numbers over time")
plt.legend()
plt.ylim([0, chart_size])
```

### Age
In the above graph, we can see that while overall student numbers have increased at The University of Birmingham, this drive has come from an increase in school leavers. Non-18-year-old student - presumably mature - numbers have remained fairly consistent.

## Ratio of Q5 to Q1 over time

### applicants

```python
offered_places_Q1 = birm_df[(birm_df.statistic=="Offers") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q1")].reset_index(drop=True)
offered_places_Q5 = birm_df[(birm_df.statistic=="Offers") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q5")].reset_index(drop=True)
offer_rate_Q1 = birm_df[(birm_df.statistic=="Offer rate") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q1")].reset_index(drop=True)
offer_rate_Q5 = birm_df[(birm_df.statistic=="Offer rate") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q5")].reset_index(drop=True)
placed_rate_Q1 = birm_df[(birm_df.statistic=="All placed applicants per 10,000 population") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q1")].reset_index(drop=True)
placed_rate_Q5 = birm_df[(birm_df.statistic=="All placed applicants per 10,000 population") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q5")].reset_index(drop=True)
placed_Q1 = birm_df[(birm_df.statistic=="All placed applicants") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q1")].reset_index(drop=True)
placed_Q5 = birm_df[(birm_df.statistic=="All placed applicants") & (birm_df.agegroup=="18 year olds") & (birm_df.equality_dimension=="POLAR4 Q5")].reset_index(drop=True)
```

```python
offered_places_Q1['value'][0]
```

```python
POLAR4_ratio_offers = []
POLAR4_ratio_offer_rate = []
POLAR4_ratio_placed = []
POLAR4_ratio_placed_rate = []
for i in range(0, len(offered_places_Q1)):
    POLAR4_ratio_offers.append(offered_places_Q5['value'][i]/offered_places_Q1['value'][i])
    POLAR4_ratio_offer_rate.append(offer_rate_Q5['value'][i]/offer_rate_Q1['value'][i])
    POLAR4_ratio_placed.append(placed_rate_Q5['value'][i]/placed_rate_Q1['value'][i])
    POLAR4_ratio_placed_rate.append(placed_Q5['value'][i]/placed_Q1['value'][i])
```

```python
POLAR4_ratio_placed_rate
```

```python
## POLAR4 comparison
#present data as line graph
x = years
plt.figure(figsize=(13,8))
plt.bar(x, POLAR4_ratio_offers, color="red", label="Total")




# Create a legend for our chart


plt.grid()
plt.xlabel("Year")
plt.ylabel("Q5/Q1")
plt.title(f"Ratio of offers between POLAR4 Q5 and Q1")
plt.legend()
```

ratio of Q5:Q1 offers has halved from ~ 8:1 to ~4:1 in eleven years. This suggests the UoB's intake is becoming more equitable.

```python
## POLAR4 comparison
#present data as line graph
x = years
plt.figure(figsize=(13,8))
plt.bar(x, POLAR4_ratio_offer_rate, color="red", label="Total")




# Create a legend for our chart


plt.grid()
plt.xlabel("Year")
plt.ylabel("Q5/Q1")
plt.title(f"Ratio of offers - adjusted for population - between POLAR4 Q5 and Q1")
plt.legend()
```

Once adjusted for population, we see that the offer rate between Q5 and Q1 is actually fairly consistent.

```python
## POLAR4 comparison
#present data as line graph
x = years
plt.figure(figsize=(13,8))
plt.bar(x, POLAR4_ratio_placed, color="orange", label="Total")




# Create a legend for our chart

plt.rcParams['axes.facecolor'] = 'whitesmoke'

plt.grid()
plt.xlabel("Year")
plt.ylabel("Q5/Q1")
plt.title(f"Ratio of placed students between POLAR4 Q5 and Q1")
```

Actual students placed has increased from a ~7:1 ratio to a ~3:1 ratio, comparing Q5 to Q1. This suggests the UoB's intake is becoming more equitable.

```python
## POLAR4 comparison
#present data as line graph
x = years
plt.figure(figsize=(13,8))
plt.bar(x, POLAR4_ratio_placed_rate, color="orange", label="Total")



plt.rcParams['axes.facecolor'] = 'whitesmoke'

# Create a legend for our chart


plt.grid()
plt.xlabel("Year")
plt.ylabel("Q5/Q1")
plt.title(f"Ratio of placements - adjusted for population - between POLAR4 Q5 and Q1")
```

Rate of students placed has increased, comparing Q5 to Q1, from a ~8:1 ratio to a ~4:1 ratio. This again suggests the UoB's intake is becoming more equitable.

This looks like very good news - by this measure, moving from a ratio of 8:1 to a ratio of 3:1 in favour of Q5 to Q1, in eleven years, is impressive. In that time, the ratio of applications to UoB has changed by a similar ratio.

Also, the fact that Q5 send four times as many applications to UoB (2021) is reflected in offers coming out at the same ratio of 4:1. However, the final placement ratio is an even more favourable 3:1. In other words, while the university is favouring Q5 students four times as much as Q1 students in terms of offers, it is less favourable than this when comparing which students actually end up attending UoB.

However, POLAR4 is regarded as a poor measure of disadvantage; it only correlates 0.22 with deprivation and 0.38 with income overall. It biases against BAME people, London residents, young mothers, and renters.
